```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv('/content/mtcars.csv')
print(data.head())
plt.scatter(data['wt'],data['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.title('scatter plot')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv('/content/mtcars.csv')
print(data.head())
plt.hist(data['mpg'], bins=10)
plt.xlabel('weight')
plt.ylabel('mpg')
plt.title('histogram')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv('/content/mtcars.csv')
print(data.head())
sv=data['mpg'].value_counts()
sv.plot(kind='bar')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.title('Frequency dist table')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv('/content/mtcars.csv')
plt.boxplot(data['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.title('box plot')
plt.grid(True)
plt.show()
```
